# Loading expense data in Jupyter

You can open your expense data in two ways: from the **CSV file** or directly from the **SQLite database**.

For ML later you'll typically use the CSV (or a pandas DataFrame).

## 1. Export to CSV (run this from the command line first)

From the project folder run once (or whenever you want a fresh export):

```bash
cd python-project
python -m expense_tracker.export_to_csv
```

This creates `expense_tracker/expenses.csv`.

## 2. Load from CSV in the notebook

````python
import pandas as pd

# Path when notebook is inside expense_tracker/
df = pd.read_csv("expenses.csv")
df
````

If you opened Jupyter from the **project root** (`python-project/`), use:

````python
import pandas as pd

df = pd.read_csv("expense_tracker/expenses.csv")
df
````

## 3. Or load directly from SQLite (no CSV needed)

This reads the latest data from the database. Path is from the notebook's working directory.

````python
import pandas as pd
import sqlite3

# When notebook is inside expense_tracker/
conn = sqlite3.connect("expenses.db")
df = pd.read_sql_query("SELECT * FROM expenses ORDER BY date DESC", conn)
conn.close()
df
````

From project root:

````python
import pandas as pd
import sqlite3

conn = sqlite3.connect("expense_tracker/expenses.db")
df = pd.read_sql_query("SELECT * FROM expenses ORDER BY date DESC", conn)
conn.close()
df
````
